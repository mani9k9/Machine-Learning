# Project 5: Customer Segmentation with Clustering 🛍️

This notebook performs customer segmentation using K-Means clustering and hierarchical clustering.

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
sns.set(style='whitegrid')
```

```python
df = pd.read_csv('/content/Datasets/Mall Dataset/Mall_Customers.csv')
df.head()
```

```python
df.drop('CustomerID', axis=1, inplace=True)
df.describe()
```

```python
# Gender vs Spending
sns.boxplot(x='Gender', y='Spending Score (1-100)', data=df)
plt.show()
```

```python
# Income vs Spending Clustering
X1 = df[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
X1_scaled = scaler.fit_transform(X1)

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X1_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Income_Cluster'] = kmeans.fit_predict(X1_scaled)

sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Income_Cluster', data=df)
plt.show()
```

```python
# Feature Engineering
df['Spending_Income_Ratio'] = df['Spending Score (1-100)'] / df['Annual Income (k$)']
df.head()
```

```python
# Hierarchical Clustering
plt.figure(figsize=(10,6))
sch.dendrogram(sch.linkage(X1_scaled, method='ward'))
plt.show()
```
